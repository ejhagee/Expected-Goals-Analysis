# Expected Goals Analysis

## Description

Expected Goals (xG) is the most popular advanced metric for association football (soccer) analysis.  In it's most basic form, it is a measurement of the probability of a particular shot resulting in a goal.  Variables taken into account include distance and angle of shot, as well as the part of the body used to take the shot.  Non-shot xG models can also be developed that perform the same task, but measure expected goals based on actions such as interceptions near goal.  In any case, expected goals for a game and/or season can be summed for a description of team performance besides the scorelines.

In this analysis, we look at the predicted (xG) versus actual (G) results across various professional leagues.

## Data

Unfortunately, due to the work required to calculate the variables for a shot, typically involving video of the shot in question, expected goal data is not widely available.  However, some free public sources exist.

Fivethirtyeight.com (538) uses two expected goals models (shot-based and non-shot based) as part of its soccer league forecasting.  The data for summed expected goals models (as well as actual results) is published periodically game-by-game.  This data is used as a first start for investigation.  As expected goals models can have a decently high variance game-by-game, we sum up for individual seasons in any case.

### Seasons

Several issues come up with labeling and grouping seasons.

First, while most countries in this dataset have Fall-Spring calendars, some have Spring-Fall calendars, due to weather and commerical concerns (namely leagues in China, Japan, Norway, Sweden, and North America).

Also, while Mexico follows a Fall-Spring calendar, it holds two tourneys in this calendar, the Apertura and the Clausura (Fall/Spring) with championships for each.  The schedule (home/away) reverses for the regular seasons.  For ease of value, we will combine these into one season.

```python
#combine Mexican Primera Division Tourneys
df538.loc[df538['league'].str.contains('Mexican Primera'), 'league'] = 'Mexican Primera Division'
```

```python
#check leagues list
df538['league'].unique()
```

Now, we can move on to labeling the calendar.

```python
#save list of spring-fall leagues
spring_start = ['Major League Soccer', 'United Soccer League', "National Women's Soccer League", 'Brasileiro Série A',
               'Swedish Allsvenskan', 'Norwegian Tippeligaen', 'Chinese Super League', 'Japanese J League']
```

```python
#label calendar used
df538['calendar'] = ['Spring-Fall' if lg in spring_start else 'Fall-Spring' for lg in df538['league']]
```

```python
#verify by value counts
df538['calendar'].value_counts()
```

```python
#create a function to label the season
def season(row):
    #check calendar
    if row["calendar"] == 'Fall-Spring':
        #check date
        if row["date"].month < 7:
            return str(row["date"].year - 1) + '-' + str(row["date"].year - 2000)
        else:
            return str(row["date"].year) + '-' + str(row["date"].year + 1 - 2000)
    else:
        return str(row["date"].year)
```

```python
#add season
df538 = df538.assign(season = df538.apply(season, axis = 1))
```

### Summarize a season

Now, we can practice the aggregation of data for one season to build up operations that we can turn into a function.

```python
#select English (Barclays) Premier League
prem_lg_df = df538.loc[df538['league'] == 'Barclays Premier League', :]
#select most recent season
test = prem_lg_df.loc[prem_lg_df['season'] == '2017-18']
#preview
test.head()
```

Teams play both at home (as team1) and away (as team2), so we need to do two groupby's.

```python
home = test.groupby('team1')
away = test.groupby('team2')
```

We want to create a dataframe with games played, goals for, goals against, and the same for all the expected goals values for the various models. 

```python
#home statistics
home_g = home['score1'].count()
home_gf = home['score1'].sum()
home_ga = home['score2'].sum()
home_xgf = home['xg1'].sum()
home_xga = home['xg2'].sum()
home_nsxgf = home['nsxg1'].sum()
home_nsxga = home['nsxg2'].sum()
home_agf = home['adj_score1'].sum()
home_aga = home['adj_score2'].sum()

#away statistics
away_g = away['score2'].count()
away_gf = away['score2'].sum()
away_ga = away['score1'].sum()
away_xgf = away['xg2'].sum()
away_xga = away['xg1'].sum()
away_nsxgf = away['nsxg2'].sum()
away_nsxga = away['nsxg1'].sum()
away_agf = away['adj_score2'].sum()
away_aga = away['adj_score1'].sum()
```

```python
#summary statistics
g = home_g + away_g
home_gd = home_gf - home_ga
away_gd = away_gf - away_ga
home_xgd = home_xgf - home_xga
away_xgd = away_xgf - away_xga
home_nsxgd = home_nsxgf - home_nsxga
away_nsxgd = away_nsxgf - away_nsxga
home_agd = home_agf - home_aga
away_agd = away_agf - away_aga
gf = home_gf + away_gf
ga = home_ga + away_ga
xgf = home_xgf + away_xgf
xga = home_xga + away_xga
nsxgf = home_nsxgf + away_nsxgf
nsxga = home_nsxga + away_nsxga
agf = home_agf + away_agf
aga = home_aga + away_aga
gd = home_gd + away_gd
xgd = home_xgd + away_xgd
nsxgd = home_nsxgd + away_nsxgd
agd = home_agd + away_agd
```

```python
#per-game-statistics
home_gfpg = home_gf/home_g
home_gapg = home_ga/home_g
home_gdpg = home_gd/home_g
home_xgfpg = home_xgf/home_g
home_xgapg = home_xga/home_g
home_xgdpg = home_xgd/home_g
home_nsxgfpg = home_nsxgf/home_g
home_nsxgapg = home_nsxga/home_g
home_nsxgdpg = home_nsxgd/home_g
home_agfpg = home_agf/home_g
home_agapg = home_aga/home_g
home_agdpg = home_agd/home_g

away_gfpg = away_gf/away_g
away_gapg = away_ga/away_g
away_gdpg = away_gd/away_g
away_xgfpg = away_xgf/away_g
away_xgapg = away_xga/away_g
away_xgdpg = away_xgd/away_g
away_nsxgfpg = away_nsxgf/away_g
away_nsxgapg = away_nsxga/away_g
away_nsxgdpg = away_nsxgd/away_g
away_agfpg = away_agf/away_g
away_agapg = away_aga/away_g
away_agdpg = away_agd/away_g

gfpg = gf/g
gapg = ga/g
gdpg = gd/g
xgfpg = xgf/g
xgapg = xga/g
xgdpg = xgd/g
nsxgfpg = nsxgf/g
nsxgapg = nsxga/g
nsxgdpg = nsxgd/g
agfpg = agf/g
agapg = aga/g
agdpg = agd/g
```

```python
#create a dataframe 
season_df = pd.DataFrame({'Team': home['team1'].first(), 'Games': g, 
                         'Home Games': home_g, 'Away Games': away_g, 'Goals Per Game': gfpg, 
                         'Goals Against Per Game': gapg, 'Goal Difference Per Game': gdpg, 
                        'Expected Goals Per Game': xgfpg, 'Expected Goals Against Per Game': xgapg,
                        'Expected Goal Difference Per Game': xgdpg, 'Non-Scoring Expected Goals Per Game': nsxgfpg,
                        'Non-Scoring Expected Goals Against Per Game': nsxgapg, 
                         'Non-Scoring Expected Goal Difference Per Game': nsxgdpg, 'Adjusted Goals Per Game': agfpg,
                        'Adjusted Goals Against Per Game': agapg, 'Adjusted Goal Difference Per Game': agdpg,
                        'Home Goals Per Game': home_gfpg, 
                         'Home Goals Against Per Game': home_gapg, 'Home Goal Difference Per Game': home_gdpg, 
                        'Home Expected Goals Per Game': home_xgfpg, 'Home Expected Goals Against Per Game': home_xgapg,
                        'Home Expected Goal Difference Per Game': home_xgdpg, 
                         'Home Non-Scoring Expected Goals Per Game': home_nsxgfpg,
                        'Home Non-Scoring Expected Goals Against Per Game': home_nsxgapg, 
                         'Home Non-Scoring Expected Goal Difference Per Game': home_nsxgdpg, 
                         'Home Adjusted Goals Per Game': home_agfpg,
                        'Home Adjusted Goals Against Per Game': home_agapg, 
                         'Home Adjusted Goal Difference Per Game': home_agdpg, 'Away Goals Per Game': away_gfpg, 
                         'Away Goals Against Per Game': away_gapg, 'Away Goal Difference Per Game': away_gdpg, 
                        'Away Expected Goals Per Game': away_xgfpg, 'Away Expected Goals Against Per Game': away_xgapg,
                        'Away Expected Goal Difference Per Game': away_xgdpg, 
                         'Away Non-Scoring Expected Goals Per Game': away_nsxgfpg,
                        'Away Non-Scoring Expected Goals Against Per Game': away_nsxgapg, 
                         'Away Non-Scoring Expected Goal Difference Per Game': away_nsxgdpg, 
                         'Away Adjusted Goals Per Game': away_agfpg,
                        'Away Adjusted Goals Against Per Game': away_agapg, 
                         'Away Adjusted Goal Difference Per Game': away_agdpg})

season_df.head()
```

```python
fig, ax = plt.subplots()
ax.scatter(season_df['Expected Goals Per Game'], season_df['Goals Per Game'], label = 'All', c = 'black')
ax.scatter(season_df['Home Expected Goals Per Game'], season_df['Home Goals Per Game'], label = 'Home', c = 'blue')
ax.scatter(season_df['Away Expected Goals Per Game'], season_df['Away Goals Per Game'], label = 'Away', c = 'red')
ax.plot([0.7, 2.5], [0.7, 2.5], c = 'orange')
ax.set_title('2017-18 English Premier League')
ax.set_xlabel('Expected Goals')
ax.set_ylabel('Goals')
ax.legend()
plt.show()
```
